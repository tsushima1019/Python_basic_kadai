```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}（年齢: {self.age}）は大人です")
        else:
            print(f"{self.name}（年齢: {self.age}）は大人ではありません")

# Humanインスタンスを複数作成
humans = [
    Human("山田太郎", 25),
    Human("佐藤花子", 18),
    Human("鈴木一郎", 30),
    Human("田中次郎", 15)
]

# リストの全インスタンスでcheck_adultメソッドを呼び出す
for person in humans:
    person.check_adult()
```
